# COMPARE VALUES OF DATA WITH PYTHON CODE

## Conditions

- Equals: x == y
- Not Equals: x != y
- Less than: x < y
- Less than or equal to x <= y
- Greater than: x > y
- Greater than or equal: x >= y


```python
temp = 50
print(temp >= 32)
print(temp < 32)
print(temp == 32)
```

```python
#Q1
rock = "basalt"
print("highland" == rock)
print("basalt" == rock)
```

```python
#Q2
rock = "basaltrock"
print("highland" in rock)
print("basalt" in rock)
```

# CONTROL THE FLOW OF THE CODE USING CONDITIONAL STATEMENTS

## if Statement

if(conditionToTestIsTrue):
   doSomething

```python
basalt = 0
if(basalt == 0):
    print("We have found no basalt rocks.")
print("Done checking basalt rocks.")
```

```python
basalt = 1
if(basalt == 0):
    print("We have found no basalt rocks.")
print("Done checking basalt rocks.")
```

## else statement

```python
basalt = 0
if(basalt == 0):
    print("We have found no basalt rocks.")
else:
    print("We found some basalt rocks!")
print("Done checking basalt rocks.")
```

```python
basalt = 5
if(basalt == 0):
    print("We have found no basalt rocks.")
else:
    print("We found some basalt rocks!")
print("Done checking basalt rocks.")
```

## Else-if statement

```python
basalt = 5
if(basalt == 0):
    print("We found no basalt rocks.")
elif(basalt == 1):
    print("We found exactly 1 basalt rock.")
else:
    print("We found more than 1 basalt rock!")
print("Done checking basalt rocks.")
```

# REPEAT CODE WITH WHILE AND FOR LOOPS

## While Loops

```python
countdown = 5

while countdown >= 0:
    print(countdown)
    countdown = countdown - 1  
print("Lift Off")
```

```python
# Trying to find life outside our planet
timer = 10
while  timer < 0:
    print("Hello, I am from Earth")
```

## For Loops

```python
planets = "Mars", "Saturn", "Jupiter"

for planet in planets:
    print(planet)
```

# USE PARAMETERS TO SPECIFY VARIABILITY IN FUNCTIONS

```python
def OutputRocketText():
    rocketNumber = 1
    print("Rocket will launch soon!")

    return

OutputRocketText()
print(rocketNumber)
```

```python
rocketText = "We will launch in"
def OutputRocketText():
    rocketText = rocketText + " two days"
    return

OutputRocketText()
```

```python
rocketText = "We will launch in"

def OutputRocketText():
    global rocketText
    rocketText = rocketText + " two days"
    return

OutputRocketText()
print(rocketText)
```

```python
def OutputRocketText(textInput):
    textInput = textInput + " two days"
    return textInput

rocketText = "We will launch in"
newRocketText = OutputRocketText(rocketText)
print(newRocketText)
```
